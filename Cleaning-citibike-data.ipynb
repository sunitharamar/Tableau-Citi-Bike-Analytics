```python
# Using this to concatenate the csv as it is much faster.

import glob, os 
import pandas as pd
import sys

path = r"./Data/**"
print(path)
all_rec = glob.iglob(os.path.join(path, "*.csv"), recursive=True)    
dataframes = (pd.read_csv(f) for f in all_rec)
print(dataframes)
big_dataframe = pd.concat(dataframes, ignore_index=True)
print(big_dataframe)
```
